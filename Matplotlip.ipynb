```python
import numpy as np
x = np.linspace(0,5,11)
y= x**2
```

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
#plt.show() # if you're not in jupyter notebook
```

```python
plt.subplot(1,2,1) # 1 = # of row, 2 is # of column and the 1 means first plot
plt.plot(x,y,'r')

plt.subplot(1,2,2) # 1 = # of row, 2 is # of column and the 1 means second plot
plt.plot(y,x,'b')
```

### Object Oriented

```python
### Object oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # add_axes takes 4 argument which are left, bottom and wirght and hight which are percent of arguments
##with OO we have more control creating chart

axes.plot(x,y) # functional creating, less contorl, it defines the axes, etc
axes.set_xlabel("X Label")
axes.set_ylabel("Y LAbel")
axes.set_title("Set Title")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('Xlabel')
axes.set_ylabel('YLabel')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

## Part2

```python
fig, axes= plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
#axes.plot(x,y)
```

```python
axesfig, axes= plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2)) #to define the size of the chart

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols = 1, figsize= (8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)
plt.tight_layout
```

```python
fig.savefig('my_picture.png', dpi=200)
```

```python
#other version

fig = plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x**2')
ax.plot(x,x**3, label ='x**3')

ax.legend(loc=0)
```

### Part3

```python
# Plot Appearance

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', linewidth=3, alpha = 0.5) #any rgb cood for costum colrs
```

```python

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', lw=3,linestyle = '--')
```

```python

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', lw=3,ls = '-')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', lw=3,ls = '-', marker ='*', markersize=20,
       markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', lw=3,ls = '--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
